```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("73087_000AD.csv", sep=";")
df.head()
colonnes_interessantes = ["adresse_nom_voie", "valeur_fonciere", "type_local", "surface_reelle_bati","date_mutation"]
df_interessant = df[colonnes_interessantes]
print(df_interessant.head())
#df_interessant.plot()
# tendances du march
df['date_mutation'] = pd.to_datetime(df['date_mutation'], format='%Y-%m-%d')
df['annee'] = df['date_mutation'].dt.year
donnees_par_annee = df.groupby('annee')['valeur_fonciere'].mean()
plt.figure(figsize=(12, 6))
plt.plot(donnees_par_annee.index, donnees_par_annee.values, marker='o', linestyle='-')
plt.title("volution des valeurs foncires par anne")
plt.xlabel("Anne")
plt.ylabel("Valeur Foncire Moyenne")
plt.grid(True)
plt.show()
donnees_par_annee.plot()
```
